
# 1. Create a pandas DataFrame using the employee data provided below

```python
import pandas as pd

data = [
    {"EmployeeID": 201, "Name": "Sarah",    "Department": "Engineering", "Salary": 85000, "Performance Score": 4, "Years of Service": 5},
    {"EmployeeID": 202, "Name": "Micheal",  "Department": "Sales",       "Salary": 62000, "Performance Score": 3, "Years of Service": 3},
    {"EmployeeID": 203, "Name": "Jennifer", "Department": "Marketing",   "Salary": 71000, "Performance Score": 5, "Years of Service": 7},
    {"EmployeeID": 204, "Name": "David",    "Department": "Engineering", "Salary": 92000, "Performance Score": 5, "Years of Service": 8},
    {"EmployeeID": 205, "Name": "Lisa",     "Department": "HR",          "Salary": 58000, "Performance Score": 4, "Years of Service": 4},
    {"EmployeeID": 206, "Name": "James",    "Department": "Sales",       "Salary": 68000, "Performance Score": 4, "Years of Service": 6},
    {"EmployeeID": 207, "Name": "Emma",     "Department": "Engineering", "Salary": 78000, "Performance Score": 3, "Years of Service": 2},
    {"EmployeeID": 208, "Name": "Robert",   "Department": "Marketing",   "Salary": 75000, "Performance Score": 4, "Years of Service": 5},
    {"EmployeeID": 209, "Name": "Sophia",   "Department": "Sales",       "Salary": 65000, "Performance Score": 5, "Years of Service": 4},
    {"EmployeeID": 210, "Name": "William",  "Department": "HR",          "Salary": 60000, "Performance Score": 3, "Years of Service": 3}
]

df = pd.DataFrame(data)
```

# 2. Select and display only the 'Name', 'Department', and 'Salary' columns

```python
print(df.iloc[:, 1:4])
print(df[["Name", "Department", "Salary"]])
```

#3. Use iloc to display the employee at row position 5

```python
print(df.iloc[4])
```

#4. Use loc to display the employee with EmployeeID 208

```python
df.index = df['EmployeeID']
print(df.loc[208])
```

#5. Filter and display all employees from the 'Engineering' department

```python
print(df.loc[df["Department"] == "Engineering"])
```

#6. Filter and display all employees with Salary greater than 70000

```python
print(df.loc[df["Salary"] > 70000])
```

#7. Filter and display employees from 'Sales' department with Performance Score >= 4

```python
print(df.loc[(df["Department"] == "Sales") & (df["Performance Score"] >= 4)])
```

#8. Sort the entire DataFrame by Salary in descending order and display the top 3 highest-paid employees

```python
print(df.sort_values("Salary", ascending=False).head(3)[["EmployeeID", "Name", "Department", "Salary"]])
```

#9. Sort by Department (ascending) and then by Salary (descending) within each department

```python
print(df.sort_values(by=["Department", "Salary"], ascending=[True, False])[["EmployeeID", "Name", "Department", "Salary"]])
```
